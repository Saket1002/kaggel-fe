```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression

df = pd.read_csv("./data/ames.csv")
```

```python
# helper function


def score_dataset(X, y, model):
    for column in X.select_dtypes(include="object").columns:
        X[column], _ = X[column].factorize()
    # Metric for Housing competition is RMSLE (Root Mean Squared Log Error)
    score = cross_val_score(
        model, X, y, cv=5, scoring="neg_mean_squared_log_error",
    )
    score = -1 * score.mean()
    score = np.sqrt(score)
    return score
```

```python
# create baseline

X = df.copy()
y = X.pop("SalePrice")


model_1 = LinearRegression()
model_2 = XGBRegressor()

print("baseline score using", type(model_1), score_dataset(X, y, model_1))
print("baseline score using", type(model_2), score_dataset(X, y, model_2))
```

```python
# lets cluster data based on year of built

kmeans = KMeans(n_clusters=3)
# note label y is not supplied to clusteing algo netiher as sepearate param nor as part of X.
cluster = kmeans.fit_predict(X['YearBuilt'].to_numpy().reshape(-1,1))
X['YearBuilt_cluster'] = cluster
```

```python
df['YearBuilt_cluster'] = cluster
print("cluster centroid years", kmeans.cluster_centers_)
sns.lmplot(x="YearBuilt", y="SalePrice", hue="YearBuilt_cluster", height=4, aspect=2, data=df);
```

```python

model_1 = LinearRegression()
model_2 = XGBRegressor()

print("with YearBuilt_cluster score using", type(model_1), score_dataset(X, y, model_1))
print("with YearBuilt_cluster score using", type(model_2), score_dataset(X, y, model_2))
```

                                LinearRegression                XGBRegressor
    
    Baseline                    0.19311846124737714             0.14284209833401057
    with YearBuilt_cluster      0.19299351903760503             0.14284209833401057


observation: YearBuilt_cluster has no impact on XGBoost however slight improvement (error reduction in case of LinearRegression)



Above was an example of very simple clustering lets try a bit more complecated.

lets build a cluster using multile area based cluster

features: LotArea, TotalBsmtSF, FirstFlrSF, SecondFlrSF,GrLivArea

number of clusters: 10


```python
df = pd.read_csv("./data/ames.csv")

X = df.copy()
y = X.pop("SalePrice")


# YOUR CODE HERE: Define a list of the features to be used for the clustering
features = ["LotArea", "TotalBsmtSF", "FirstFlrSF", "SecondFlrSF","GrLivArea"]


# Standardize
X_clustering_features = X.loc[:, features]
X_scaled_clustering_features = (X_clustering_features - X_clustering_features.mean(axis=0)) / X_clustering_features.std(axis=0)


# YOUR CODE HERE: Fit the KMeans model to X_scaled and create the cluster labels
kmeans = KMeans(n_clusters=10, n_init=10, random_state=0)
X["area_cluster"] = kmeans.fit_predict(X_scaled_clustering_features)
df["area_cluster"] = X["area_cluster"]
```

```python

model_1 = LinearRegression()
model_2 = XGBRegressor()

print("with area_cluster score using", type(model_1), score_dataset(X, y, model_1))
print("with area_cluster score using", type(model_2), score_dataset(X, y, model_2))
```

                                LinearRegression                XGBRegressor
    
    Baseline                    0.19311846124737714             0.14284209833401057
    with YearBuilt_cluster      0.19299351903760503             0.14284209833401057
    area_cluster                0.1933485135277746              0.1425257903378395

observation: YearBuilt_cluster has no impact on XGBoost however slight improvement (error reduction in case of LinearRegression)

observation: with area_cluster linear regresison error went up while for XGBRegressor it went down. 

## so while impact is minumum it has an important take away FE is model specific, they impact different model differently.



```python
df["med_house_price"] = df.groupby("area_cluster")["SalePrice"].transform("mean")
```

```python
sns.lmplot(x="area_cluster", y="med_house_price", height=4, aspect=2, data=df)
```

```python
X['med_house_price'] = df["med_house_price"]

model_1 = LinearRegression()
model_2 = XGBRegressor()

print("with area_cluster score using", type(model_1), score_dataset(X, y, model_1))
print("with area_cluster score using", type(model_2), score_dataset(X, y, model_2))
```
